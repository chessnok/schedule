# Постановка задачи

## Входные данные
### Массивы
- $ D = \begin{array}{cc}
     A & B
\end{array} $ для онсайта, или числа от 1 до 5 для онлайна - дни недели (day).
- $ LOC $ список всех мест для занятий.
- $ L_t $ список мест, где проводит занятия учитель t, где $ L_t_i \in LOC $.
- $ P = 1, 2, \ldots, 7 $ для онсайта, или от 1 до n для онлайна - слоты времени (slot).
- $ T $ список учителей, таких как artem, nastya и т. д.
- $ S $ - список учеников.
- $ G $ - список групп по учителям, где $ i \in G_t $ значит, что у учителя t есть группа i.
- $ SG_t_g $ где $ g \in G_t $ - список учеников по группам, где $ s \in SG_t_g $. Это означает, что в группе g, относящейся к учителю t, есть ученик s.
- $ C_t_g $ где $ t \in T $ - количество занятий в неделю у группы g учителя t.
- $ GS_(ts) $ где $ s \in S $ - список пар (учитель+группа) для ученика s.
- $ GC_{tg} $ где $ g \in G $ - где $ i \in GC_g $ означает, что группа g учителя t не может проводить занятия в один таймслот с i. = все остальные группы учителя t + все остальные группы всех учеников группы g.
### Константы
- min_free_teachers = минимальное кол-во свободных учителей в течение одного периода
- teachers_constant = константа, которая показывает насколько важно минимизировать кол-во уроков в день для учителей, вместо учеников. По умолчанию len(T) / len(S)
### Параметры

- $ d \in D $
- $ p \in P $
- $ t \in T $
- $ l \in LOC \cup G_t $
- $ g \in G_t $
$ X_{dptlgs} = 1 $ если учитель t проводит занятие в месте l в день d в слоте времени p для группы g и ученика s.
$ Y_{dptgl} = 1 $ если в месте l в день d в слоте времени p проводится занятие группе g.

### Ограничения
- Занятие t g не мог проводиться парралельно с 
- В одном помещении проводиться только одно занятие
- Каждый слот свободно как минимум min_free_teachers учителей

### Целевая функция
- Минимизировать количество уроков в день (важно как для учителей, так и для учеников), но для учителей * teachers_constant
- Optional: минимизировать различия с текущим расписанием
